## Activation Functions
* In our neural network whenever we are going to send the dataset so in a hidden layer there will be an activation function.
* Activation function behaves like a gate and it will try to normalize my dataset. 
* Some activation function will try to give a data in between 0 to 1. and some activation function will give 1 to -1. 

### 1.Sigmod Function

* This activation function is one of the commonly used function.
* Whatever input we are going to calculate based on (wx+b), the output of this will give into the x .
* In this graph it tries to fluctuate in between 0 to 1.
* Whatever input we will give, the value of output always ranges between 0 to 1 and it's not give anyother output except 0 and 
* This sigmoid(0 to 1) is not a kind of a probability function.

#### Advantages:
1. In sigmoid function we will be able to see smooth gradient.
2. The value of output always ranges between 0 to 1
3. It has a smooth gradient 
4. It is very useful for classifiers.

#### Disadvantages:
1. Sigmoid is a sort of an exponential base function and it is a bit heavy operation.
2. It's give rise to a problem of Vanishing gradient(Value of x increases then your convergence of your gradient starts decreasing)
3. Sigmoid is not a zero centric 
4. The network refuses to learn further or it is drastically slow.

### 2.Tanh function

* Tanh is a zero centric function which means it fluctuates across zero value.
* The output of this function should always between -1 to 1.
* This function is very much helpful to normalize the datasets.
* This is mainly used for Binary class classifications.(tanh in a hidden layer and sigmoid in output layer)

#### Advantages:
1. Zero centric one.
2. Extensively used for binary class classification.
3. Smooth gradient

#### Disadvanages:
1. Vanishing gradient problem.
2. This function is bit complex for dense networks.
3. It is also a kind of exponential base function,so it is a bit heavy operation.

### 3.Relu - Rectified linear unit

* This function gives you the maximum value
* In a positive axis,any value of x it's going to give you a maximum one.
* In case of a negative axis,any value of x in the negative axis it always gives you the value 0.
* In case of a relu function there is no smooth gradient you will be able to see in the negative axis.
* The value ranges between 0 to infinity.

#### Advantages:
1. In positive x axis the major adantage will be there is no Vanishing gradeient problem and you will be able to observe gradient saturation(Keep on getting slopes).
2. Calculation speed is much faster than sigmoid and Tanh function.
3. It is a kind of Linear calculation and not an exponential base function.

#### Disadvantages:
1. Incase of a negative axis it always gives you the 0, So in that case you will not be able to see any smooth gradient in negative axis.Due to this some gradients can be fragile during training and can die.
2. Simply saying that Relu could result in Dead Neurons. 
3. Not a zero centric function.
4. This function is only be used within hidden layers of a Neural network model.


### 4.Leaky Relu - Leaky rectified linear unit

* LeakyRelu is a variant of Relu.
* Instead of being 0 when x<0(negative x axis), a LeakyRelu allows a small,non-zero,constant gradient(0.01).
* The value ranges between 0 to infinity.

#### Advantages:
1. Calculation is very simple and faster.
2. Even for the negative axis it is going to adjust the gradient.
3. LeakyRelu are one attempts to fix the Dyeing Relu by having a small negative slope.
4. No Vanishing gradient in positive x axis.
5. It is a kind of Linear calculation and not an exponential base function.

#### Disadvantages:
1. As it possess linearity,it can't be used for the complex classification.It lags behind the Sigmoid and Tanh for some of the use cases.


### 5.ELU - Exponential linear unit

* ELU is a function that tends to converge cost to zero faster and produce more accurate results.
* For this activation function, an alpha value is picked.
* Different to other activation functions, ELU has a extra alpha constant which should be positive number.
* ELU is very similar to RELU except negative inputs.

#### Advantages:
1. No gradient clipping or No Vanishing gradient.
2. The average of the output is almost close to zero.So this function is good for normalization.
3. ELU is a strong alternative to Relu.
4. Unlike to Relu,ELU can produce negative outputs
5. Does'nt have the dyeing Relu problem.

#### Disadvantages:
1. It has longer computation time,because of the exponential operation.
2. Does not avoid the exploding gradient problem
3. The Neural network doesnot learn the alpha value.
4. ELU is more computationally expensive than the Relu.

### 6.PRelu - Parametric relu

* PRelu is a type or variant of a LeakyRelu.
* Instead of having a predetermined slope like 0.01,makes it a parameter for the neural network to figure out itself y=alpha*x
  when x<0.
* If alpha is a learnable parameter which means by doing some kinds of an experiments with respect to the slope or gradient.So   in that case that is called as Parametric Relu function.
* So whatever input you will give it will try to learn out of those inputs and then for the negative axis and then it will       going to set the alpha value.
* From the PRelu itself we can derive the value and you will be able to derive the LeakyRelu.

#### Advantages:
1. Allows the negative slope to be learned.

#### Disadvantage:
1. May perform differently for different problems.



### 7.Softmax function

* Softmax function calculates the probability occurence or probability distribution of the event over n different events.
* This Softmax activation function used in outer layer where you have to findout the probability of the multiple or different     events.

#### Advantages
1. This function is mainly used for Multiclass classification.
2. In building Neural networks softmax function used in different layer level.
3. The high value will have the higher probability than other values.
4. Able to handle multiple classes.
5. Useful for output Neurons.

### 8.Swish Function

* Swish function is a self gated function.
* Self gated means that the gate is just the sigmoid of the activation itself.
* In this self gated function you will not get a Vanishing gradient problem. 
* Swish is basically inspired by the sigmoid function for getting a LSTM kind of a network or Highway kind of networks.
* Gates in the form of sigmoids that control how much of a vector gets passed on to a next stage,by multiplying it between the   output of the sigmoid,which is a number between 0 and 1.
* In LSTM network you are able to train the network weight slowly and same point of a time it will be able to memorize some       past datasets.

#### Advantages:
1. Swish function can be regarded as a smooth function.
2. This function is applicable for large and small datasets.

### 9.Maxout Function

* This function takes the maximum value among the values from n linear functions.
* Maxout is a combination of Relu and LeakyRelu
* First it will learn and out of the learnable parameter it gives you a final value and it should pass through the particular     function.
* This kind of an activation function which doesn't depends upon the predefined situation or scenario.
* This maxout will always depends upon the weight that you are going to train and the biases that you are going to train and     based on that it is going to give you the maximum of the output.

* Softplus is a smoother function compare to the Relu function.
